```python
import pandas as pd

# Load the data (assuming already loaded as df)
df = pd.read_csv("C:\Users\pk\Desktop\2020 LB Elections Kerala\Election Data Processed.csv")  # Uncomment if needed

# Create 'Strength' category based on 'Lead'
def categorize_strength(lead):
    if lead >= 500:
        return "500+"
    elif 200 <= lead < 500:
        return "200-499"
    elif 100 <= lead < 200:
        return "100-199"
    elif 50 <= lead < 100:
        return "50-99"
    elif 1 <= lead < 50:
        return "1-49"
    elif lead == 0:
        return "0"
    elif -49 <= lead < 0:
        return "-1 to -49"
    elif -99 <= lead < -49:
        return "-50 to -99"
    elif -199 <= lead < -99:
        return "-100 to -199"
    elif -499 <= lead < -199:
        return "-200 to -499"
    else:  # lead <= -500
        return "-500 or less"

df["Strength"] = df["Lead"].apply(categorize_strength)

# Recode 'Type' to 'LBType'
lbtype_map = {
    "B": "Block",
    "G": "Grama",
    "C": "Corporation",
    "D": "District",
    "M": "Municipality"
}
df["LBType"] = df["Type"].map(lbtype_map)

# Recode 'Type' to 'Tier'
def recode_tier(t):
    if t in ["G", "C", "M"]:
        return "Ward"
    elif t == "D":
        return "District"
    elif t == "B":
        return "Block"
    else:
        return "Unknown"

df["Tier"] = df["Type"].apply(recode_tier)

# Preview the new variables
df[["Lead", "Strength", "Type", "LBType", "Tier"]].head()
```
